**Practicing Matplotlib**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
xarr= np.array([[0,0],[303,100],[303,200],[203,100],[50,100]])
yarr=np.array([[0,0],[303,200],[303,200],[203,500],[50,150]])

plt.plot(xarr,yarr)
plt.plot(xarr,yarr,'o')
plt.show()
```

```python
x=np.array([2,7,4.5,2])
y=np.array([2,2,6,2])
a=np.array([5,5,1,5])
plt.plot(x,y, marker='o',color="b",linewidth=3)
plt.plot(x,a, marker='o',color="b",linewidth=3)

```

**Plot Labels**

```python
plt.plot(x,y)
plt.xlabel("X-Axis",fontdict={'family':'Arial','color':'green','size':'16'})
plt.ylabel("Y-Axis", fontdict={'family':'Arial','color':'green','size':'16'})
plt.title("Triangle", fontdict={'family':'Arial','color':'blue','size':'24','style':'italic'})
```
